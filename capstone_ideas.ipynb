```python
import pandas as pd
pd.set_option('display.max_columns', 110)
```

```python
df = pd.read_csv('complaints.csv')
```

```python
df.head()
```

```python
for col in df.columns: 
    if (df[col].nunique() < 200) or (col == 'Company'):
        display(df[col].value_counts())
```

```python
df.size
```

```python
credit_df = df[(df['Company'].str.startswith('EQUIFAX')) | \
            (df['Company'].str.startswith('TRANSUNION')) | \
            (df['Company'].str.startswith('Experian'))]
```

```python
credit_df.size
```

```python
for col in credit_df.columns: 
    if (credit_df[col].nunique() < 200) or (col == 'Company'):
        display(credit_df[col].value_counts())
```

```python
credit_df['Date received'] = pd.to_datetime(credit_df['Date received'])
```

```python
credit_df = credit_df.set_index('Date received')
```

```python
credit_df.groupby(credit_df.index.year)['Company'].count()
```

```python
credit_df.index.year.max()
```
